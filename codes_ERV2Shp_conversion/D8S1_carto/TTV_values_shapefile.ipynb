**Import packages**

```python
import geopandas as gpd
import numpy as np
```

**Read the Shapefile**

```python
D8S1_gdf = gpd.read_file('./CartoD8S1_Isa020821.shp')
```

```python
D8S1_gdf
```

**Add the new column**

```python
D8S1_gdf_cp = D8S1_gdf.copy()

D8S1_gdf_cp["Unc"] = np.where(D8S1_gdf_cp["LineType"]==0, 1.0, 0.5)
#D8S1_gdf_cp["Unc"] = np.where(D8S1_gdf_cp["LineType"]==0, 1.0, 0) #NoUnc
```

```python
D8S1_gdf_cp
```

**Save the Geo DataFrame into a Shapefile**

```python
# Assign CRS
D8S1_gdf_cp.crs = D8S1_gdf.crs

# Save gdf into a shapefile
D8S1_gdf_cp.to_file('./CartoD8S1_Isa020821_Unc.shp')
```
